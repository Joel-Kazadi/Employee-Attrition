## Employee turnover: ***workers are leaving the company***

**Attrition is a process in which the workforce dwindles at an organization, following a period in which a number of people retire or resign, and are not replaced.**
- A reduction in staff due to attrition is often called a hiring freeze and is seen as a less disruptive way to trim the workforce and reduce payroll than layoffs.
- In this NoteBook our Aim will be to analyze the datasets completely wrt each feature and find the reason behind Attrition of Employees.
- And what the top factors which lead to employee attrition?

### Description about the data

- Age: A period of employee life, measured by years from birth.
- Attrition: The departure of employees from the organization (TARGET VARIABLE).
- BusinessTravel: Did the employee travel on a business trip or not.
- DailyRate: Employee salary for the period is divided by the amount of calendar days in the period.
- Department: In which department the Employee working.
- DistanceFromHome: How far the Employee live from the office location.
- Education: In education 1 means 'Below College', 2 means 'College', 3 means 'Bachelor', 4 means 'Master', 5 means 'Doctor'
- EducationField: In which field Employee complete his education.
- EmployeeCount: How many employee working in a department
- EmployeeNumber: An Employee Number is a unique number that has been assigned to each current and former State employee and elected official in the Position and Personnel DataBase (PPDB).
- Job involvement: Is the degree to which an employee identifies with their work and actively participates in it where 1 means 'Low', 2 means 'Medium', 3 means 'High', 4 means 'Very High'
- JobLevel: Job levels, also known as job grades and classifications, set the responsibility level and expectations of roles at your organization. They may be further defined by impact, seniority, knowledge, skills, or job title, and are often associated with a pay band. It's defined here from low level (1) to high level (5).
- JobRole: What is the jobrole of an employee.
- JobSatisfaction: Employee job satisfaction rate where, 1 means 'Low', 2 means 'Medium', 3 means 'High', 4 means 'Very High'
- MaritalStatus: Marital status of the employee.
- MonthlyIncome: total monetary value paid by the organization to an employee.
- MonthlyRate: The per-day wage of the employee.
- NumCompaniesWorked: Before joining this organization how many organizations employee worked.
- Over18: Is the employee age over than 18 or not.
- OverTime: A Employee works more than 9 hours in any day or for more than 48 hours in any week.
- PercentSalaryHike: Percentage increase in an employee's salary. It's a critical factor in employee retention strategies in order to improve employee satisfaction and reduce turnover rates.
- PerformanceRating: 1 'Low' 2 'Good' 3 'Excellent' 4 'Outstanding'
- EnvironmentSatisfaction: 1 'Low' 2 'Medium' 3 'High' 4 'Very High'
- RelationshipSatisfaction: 1 'Low' 2 'Medium' 3 'High' 4 'Very High'
- StandardHours: Is the number of hours of production time that should have been used during an working period.
- StockOptionLevel:  Employee stock options, also known as ESOs, are stock options in the company’s stock granted by an employer to certain employees. Typically they are  granted to those in management or officer-level positions. Stock options give the employee the right to buy a certain amount of stock at a specific price, during a specific period of time. Options typically have expiration dates as well, by which the options must have been exercised, otherwise they will become worthless.
- TotalWorkingYears: Total years the employee working in any organization
- TrainingTimesLastYear: Last year how many times employee took training session.
- WorkLifeBalance: 1 'Bad' 2 'Good' 3 'Better' 4 'Best'
- YearsAtCompany: How many years the employee working in the current organization
- YearsInCurrentRole: How many years the employee working in the current position
- YearsSinceLastPromotion: How many years the employee working in the current position after promotion
- YearsWithCurrManager: How many years the employee working under the current manager

## Label Encodeing 

In machine learning, we usually deal with datasets that contain multiple labels in one or more than one columns. These labels can be in the form of words or numbers. To make the data understandable or in human-readable form, the training data is often labelled in words.

```python
# Target Variable (Attrition)
data['Attrition'] = data['Attrition'].replace({'No':0,'Yes':1})
```

```python
# Encode binary variables
data['OverTime'] = data['OverTime'].map({'No':0,'Yes':1})
data['Over18'] = data['Over18'].map({'N':0,'Y':1})
data['Gender'] = data['Gender'].map({'Male':0,'Female':1})
```

```python
# Encode categorical columns which are ordinal, use labelEncoding
# Apply Label encoder to df_categorical

from sklearn.preprocessing import LabelEncoder
encoding_cols=['BusinessTravel','Department','EducationField','JobRole','MaritalStatus']
label_encoders = {}
for column in encoding_cols:
    label_encoders[column] = LabelEncoder()
    data[column] = label_encoders[column].fit_transform(data[column])
```

# Resampling

Resampling is the method that consists of drawing repeated samples from the original data samples. The method of Resampling is a nonparametric method of statistical inference. In this project, we will implement hybrid sampling (SMOTEENN). This technique is a combination of both over- and undersampling techniques to balance the dataset.

```python
from imblearn.combine import SMOTEENN

resampling = SMOTEENN(sampling_strategy = "all", random_state = 12345)
features, target = resampling.fit_resample(X, y)

display(pd.DataFrame(y.value_counts(), columns = ['Before resampling']))
display(pd.DataFrame(target.value_counts(), columns = ['After resampling']))
```

The Classification Report, Confusion Matrix and Evaluation Metrics Accuracy, F1 Score and ROC-AUC Score are checked for each Classifier and are compared in order to decide which Classification Model is the best.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report, roc_auc_score,roc_curve,auc
```

```python
accuracy_list = []
f1_list = []
roc_auc_list = []
```

*Train Test Evaluation Function*

```python
def result(X, y, ts, rs, model):
    
    #train test split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=ts, random_state=rs)
    
    #scaling
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)
    
    #fit on data
    model.fit(X_train, y_train)
    
    #prediction
    pred = model.predict(X_test)
    
    #performance of model
    print("Classification Report: \n", classification_report(y_test, pred))
    print("-" * 100)
    print()
    
    #accuracy of model
    acc = accuracy_score(y_test, pred)
    accuracy_list.append(acc)
    print("Accuracy Score: ", acc)
    print("-" * 100)
    print()

    #f1-score of model
    f1 = f1_score(y_test, pred)
    f1_list.append(f1)
    print("F1 Score: ", f1)
    print("-" * 100)
    print()

    #roc-auc curve of model
    fpr,tpr,threshold = roc_curve(y_test,pred)
    auc_value = auc(fpr,tpr)
    rocauc_score = roc_auc_score(y_test, pred)
    roc_auc_list.append(rocauc_score)
    plt.figure(figsize=(5,5),dpi=100)
    print("ROC-AUC Score: ", f1)
    print("-" * 100)
    print()
    plt.plot(fpr,tpr,linestyle='-',label = "(auc_value = %0.3f)" % auc_value)
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend()
    plt.show()
    print()
    
    #confusion matrix for model
    print("Confusion Matrix: ")
    plt.figure(figsize=(10, 5))
    sns.heatmap(confusion_matrix(y_test, pred), annot=True, fmt='g');
    plt.title('Confusion Matrix', fontsize=20)
```

Logistic Regression

```python
model = LogisticRegression()
result(features, target, 0.25, 42, model)
```

Random Forest Classifier

```python
rf = RandomForestClassifier()
result(features, target, 0.25, 42, rf)
```

Gradient Boosting Classifier

```python
gb = GradientBoostingClassifier()
result(features, target, 0.25, 42, gb)
```

Naive Bayes Classifier

```python
nb = GaussianNB()
result(features, target, 0.25, 42, nb)
```

Decision Tree Classifier

```python
dt = DecisionTreeClassifier()
result(features, target, 0.25, 42, dt)
```

Stochastic Gradient Descent Classifier

```python
sgd = SGDClassifier()
result(features, target, 0.25, 42, sgd)
```

Artificial Neural Network

```python
# Building the ANN model
def build_ann_model():
    model = MLPClassifier(hidden_layer_sizes=(64, 32),  # Set the desired architecture
                          activation='relu',  # Activation function (can be 'relu', 'tanh', etc.)
                          solver='adam',  # Optimizer (can be 'adam', 'sgd', etc.)
                          max_iter=200,  # Maximum number of iterations
                          learning_rate="invscaling", # "invscaling" or "adaptive"
                          alpha=.5,
                          random_state=42)  # Random state for reproducibility
    return model

# Initializing the model
ann = build_ann_model()

# Calling the result function
result(features, target, 0.25, 42, ann)
```

```python
# !pip install pycaret==3.2.0
import pycaret
pycaret.__version__
```

```python
# creating an experimentation session

from pycaret.classification import ClassificationExperiment

data = pd.concat([features, target], axis=1)
data = data.sample(100, random_state=42)
session = ClassificationExperiment()
session.setup(data=data, target='Attrition', train_size=0.75, preprocess=True,
              data_split_stratify=True, normalize=True, fold=10, session_id=2024)

session;
```

```python
# list of available models for classification

session.models()
```

We will deploy the model trained on the Random Forest algorithm, as it showed the highest performance.

```python
# modeling on the train set

preliminary_model = session.create_model('rf')
print(preliminary_model)
```

```python
# evaluation of the preliminary model on the test set

preliminary_score = session.predict_model(preliminary_model, n_jobs=-1)
preliminary_score[['Attrition', 'prediction_label', 'prediction_score']]
```

```python
# tuning hyper-parameters

best_model = session.tune_model(preliminary_model, optimize='Accuracy')
print(best_model)
```

```python
# evaluation of the final model on the test set

final_score = session.predict_model(best_model, n_jobs=-1)
final_score[['Attrition', 'prediction_label', 'prediction_score']]
```

```python
# features importance in the model

session.plot_model(best_model, plot='feature')
pd.DataFrame({'Features': best_model.feature_names_in_,
              'Coef': best_model.coef_[0]})
```

```python
# saving the model

session.save_model(best_model, 'RF_model')
```

```python
# building the model pipeline

pipeline = session.finalize_model(best_model)
print(pipeline)
```

```python
# creating an API for deployment

turnover_api = session.create_api(pipeline, 'turnover_api')
```
